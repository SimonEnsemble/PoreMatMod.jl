# Objects

```julia
DCM = moiety("DCM")
methylene = moiety("methylene")
Ti_MIL_125_noBonds = Crystal("Ti-MIL-125.cif")
strip_numbers_from_atom_labels!(Ti_MIL_125_noBonds)
Ti_MIL_125 = deepcopy(Ti_MIL_125_noBonds)
infer_bonds!(Ti_MIL_125, true)
p_phenylene = moiety("p-phenylene")
p_tolyl = moiety("p-tolyl")
p_toluene = moiety("p-toluene")
;
```
